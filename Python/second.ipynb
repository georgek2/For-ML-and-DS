```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries = pd.read_csv('countries.csv')
countries.head()
```

```python
print(countries.tail())
print('--------------------------------')
print(countries[countries.country == 'Zambia'])
print(countries[countries.country == 'Kenya'])
```

```python
zambia = countries[countries.country == 'Zambia']
egypt = countries[countries.country == 'Egypt']
ghana = countries[countries.country == 'Ghana']
mozambique = countries[countries.country == 'Mozambique']
kenya = countries[countries.country == 'Kenya']
```

```python
plt.plot(zambia.year, zambia.population / 10**6)
plt.plot(egypt.year, egypt.population / 10**6)
plt.plot(ghana.year, ghana.population / 10**6)
plt.plot(mozambique.year, mozambique.population / 10**6)
plt.plot(kenya.year, kenya.population / 10**6)
plt.title('Population Growth in different African Countries')
plt.xlabel('Years')
plt.ylabel('Population Growth over the years')
plt.legend(['Zambia', 'Egypt', 'Ghana', 'Mozambique', 'Kenya'])
plt.show()
```

```python
plt.plot(kenya.year, kenya.population / 10**6)
plt.title('Kenya\'s population growth from 1952')
plt.xlabel('Years')
plt.ylabel('Population Increase')
plt.show()
```
