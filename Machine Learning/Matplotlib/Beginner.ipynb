```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline # Eliminates the need to keep calling the function plt.show()
```

```python
# Plotting a simple line graph..
y = [2, 4, 6, 7, 10] # A scalar variable...
plt.plot(y)
plt.show() # Assumes it is a seq of y values and generates the X values from index 0..
```

```python
plt.plot([12, 36, 48, 78], [23, 67, 123, 156], 'ro') # Specifying the color string with a line style.. red, dots..
plt.show()
```

```python
# The axis() command takes in the [xmin, xmax, ymin, ymax] and specifies the viewport of the axes. 
plt.plot([12, 36, 48, 78], [23, 67, 123, 156], 'ro')
plt.axis([11, 100, 22, 200]) # Compare this one with the above..
```

```python
# Three lines in different styles on one command
# Red dashes, blue squares and green triangles
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
```

```python
# Using Keyword strings to plot..
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# Scatter plot
plt.scatter('a', 'b', c='c', s='d', data=data) # Data keyword argument to show source
plt.xlabel('Entry a')
plt.ylabel('Entry b')
plt.show() # Turns out it has some slight difference when used
# Each run will produce a different scatter plot
```

```python
# Plotting with categorical variables
names = [['categ_a', 'categ_b', 'categ_c', 'categ_d'], [1, 20, 98, 45]]
# values = [1, 20, 98, 45]

plt.figure(figsize=(9, 3))

plt.subplot(131) # Look more into this
plt.bar(names[0], names[1]) # Plotting a bar chart of names against values
plt.subplot(132)
plt.scatter(names[0], names[1]) # Scatter Plots
plt.subplot(133)
plt.plot(names[0], names[1]) # Simple line graph

plt.suptitle('Categorical Variables Plotting')

# plt.show()
```

```python
# The use of subplots and multiple figures

plt.figure(1)              # The first figure
plt.subplot(211)        # The first subplot in it
plt.plot([4, 6, 8])
plt.subplot(212)         #The second subplot in it
plt.plot([3, 8, 5]) 
plt.title('Trial and error')

plt.figure(2)                 #The second figure
plt.plot([2, 7, 11])        # This creates a subplot(111) by default

plt.figure(1)                 # Back to fig one
plt.subplot(211)        #Referring to the subplot 211 in figure one
plt.title('Easy plotting you guy')     # Setting the title for it
plt.show()
```
